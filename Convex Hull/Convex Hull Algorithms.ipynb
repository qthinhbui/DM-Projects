# Import module

```python
import random
import time
from Jarvis_Algorithm import *
from QuickHull_Algorithm import *
from MergeHull_Algorithm import *
```

# Random Dataset

```python
random.seed(2012)
data = []
for i in range(100):
    data.append(point(random.randint(-100000000, 100000000), random.randint(-100000000, 100000000)))
```

# Jarvis Algorithm Implementation

```python
t0 = time.time()
hull_ja = convex_hull_ja(data, len(data))
t1 = time.time()
print("Runtime: {}".format(t1-t0))
print("")
display_ja(data, hull_ja)
```

# QuickHull Algorithm Implementation

```python
t2 = time.time()
hull_qh = convex_hull_qh(data, len(data))
t3 = time.time()
print("Runtime: {}".format(t3-t2))
print("")
display_qh(data, hull_qh)
```

# MergeHull Algorithm Implementation

```python
t4 = time.time()
hull_mh = convex_hull_mh(data)
t5 = time.time()
print("Runtime: {}".format(t5-t4))
print("")
display_mh(data, hull_mh)
```
