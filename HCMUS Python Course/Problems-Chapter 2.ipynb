# Bài 1

Viết chương trình cho người dùng nhập vào một khoảng cách nguyên có đơn vị là km, sau đó chuyển khoảng cách trên thành đơn vị mét (m) và in ra màn hình. Công thức chuyển đổi như sau:

Khoảng cách (km) = Khoảng cách (m) / 1000

[Dữ liệu nhập vào]

-	Một số tự nhiên là khoảng cách có đơn vị là km

[Dữ liệu xuất ra]

-	Một số tự nhiên là khoảng cách có đơn vị là mét (m) tương ứng với dữ liệu nhập.

Ví dụ:

-	Nhập vào: 10

-	Xuất ra: 10000

-	Giải thích: Chương trình được yêu cầu chuyển đổi khoảng cách nhập vào là 10km ra khoảng cách với đơn vị là m. Theo công thức chuyển đổi, ta thấy 10km x 1000 = 10000m. Chương trình xuất ra màn hình giá trị 10000.

```python
def chuyen_don_vi_km_m(khoang_cach_km):
# bat dau viet code cua ban vao duoi day
     return khoang_cach_km/1000
# ket thuc code cua ban

# Luu y: khong thay doi phan code duoi day
khoang_cach_km = int(input())
khoang_cach_m = chuyen_don_vi_km_m(khoang_cach_km)
print(khoang_cach_m)
```

-	Một số tự nhiên là tổng thu nhập kiếm được từ tiền bán vé của sân vận động.

Ví dụ (Dấu cách trong dữ liệu nhập xuất tương ứng với dấu xuống dòng):

-	Nhập vào: 10 20 5

-	Xuất ra: 4350000

-	Giải thích: Chương trình được yêu cầu tính tổng thu nhập kiếm được từ việc bán vé của sân vận động với số lượng vé bán được là 10 vé loại A, 20 vé loại B, 5 vé loại C. Tổng thu nhập được tính là 150000x10+120000x20+90000x5 = 4350000. Chương trình xuất ra màn hình giá trị 4350000.

# Bài 6

Một hành tinh hình cầu có bán kính là R. R là 1 số tự nhiên dương được nhập vào từ bàn phím. Hãy viết chương trình giúp nhận vào bán kính của hành tinh và trả về diện tích bề mặt của hành tinh đó. Công thức tính diện tích bề mặt hình cầu như sau:

S = 4 x pi x R^2 (Với số pi = 3.14)

[Dữ liệu nhập vào]

-	Một số tự nhiên R là bán kính của hành tinh

[Dữ liệu xuất ra]

-	Một số thực là diện tích bề mặt hành tinh

Ví dụ:

-	Nhập vào: 5

-	Xuất ra: 314.0

-	Giải thích: Chương trình được yêu cầu tính diện tích bề mặt hành tinh hình cầu với bán kính R = 5. Diện tích bề mặt hành tinh được tính bằng công thức 4 x 3.14 x 5^2 = 314.0. Chương trình xuất ra màn hình giá trị thực 314.0.

```python
def tinh_dien_tich_hinh_cau(R):
# bat dau viet code cua ban vao duoi day
    pi = 3.14
    return 4*pi*R**2
# ket thuc code cua ban

# Luu y: khong thay doi phan code duoi day
R = int(input())
print(tinh_dien_tich_hinh_cau(R))
```

# Bài 7

Cho a, b, k là 3 số tự nhiên (a và b không đồng thời bằng 0, k là một số dương).

Viết chương trình kiểm tra xem k có phải là ước chung của a và b hay không? Nếu đúng hãy xuất ra True. Ngược lại, chương trình xuất ra chuỗi False.

Nhắc lại định nghĩa: Số tự nhiên n chia hết cho số tự nhiên m thì ta nói n là bội của m, còn m là ước của n. Nếu số k vừa là ước của số a và vừa là ước của số b thì ta nói k là ước số chung của a và b.

[Dữ liệu nhập vào]

-	Ba dòng, mỗi dòng là một số tự nhiên lần lượt là các giá trị tương ứng với các tham số a, b, k

[Dữ liệu xuất ra]

-	Kết quả kiểm tra k có phải là ước chung của a và b hay không? Nếu đúng xuất ra chuỗi True, ngược lại, xuất ra chuỗi False.

Ví dụ (Dấu cách trong dữ liệu nhập xuất tương ứng với dấu xuống dòng):

-	Nhập vào: 12 30 3

-	Xuất ra: True

-	Giải thích: Chương trình được yêu cầu kiểm tra xem 3 có phải ước chung của 12 và 30 không? Ta thấy 12 và 30 đều chia hết cho 3, nên 3 là ước chung của 12 và 30. Chương trình xuất ra màn hình kết quả True.

```python
def kiem_tra_uoc_chung(a, b, k):
# bat dau viet code cua ban vao duoi day
    return (a % k == 0) and (b % k == 0)
# ket thuc code cua ban

# Luu y: khong thay doi phan code duoi day
a = int(input())
b = int(input())
k = int(input())
ket_qua_kiem_tra = kiem_tra_uoc_chung(a, b, k)
print(ket_qua_kiem_tra)
```

An có n viên kẹo, An định chia cho m người em mỗi người k viên kẹo. Viết chương trình tính toán xem sau khi chia kẹo, An còn lại bao nhiêu viên kẹo. Giả sử An có đủ kẹo để chia cho mỗi người em ít nhất 1 viên.

[Dữ liệu nhập vào]

-	Dòng thứ hai là số tự nhiên m là số người mà An chia kẹo.

-	Dòng thứ ba là số tự nhiên k là số kẹo mà An chia cho mỗi người.

[Dữ liệu xuất ra]

-	Một số tự nhiên là số kẹo còn lại sau khi An chia kẹo.

Ví dụ (Dấu cách trong dữ liệu nhập xuất tương ứng với dấu xuống dòng):

-	Nhập vào: 10 9 1

-	Xuất ra: 1

-	Giải thích: Chương trình được yêu cầu tính xem An còn lại bao nhiêu viên kẹo nếu An chia 10 viên kẹo của mình cho 9 người em, mỗi người 1 viên kẹo. Dễ thấy lúc này An còn lại 1 viên kẹo. Chương trình xuất ra màn hình giá trị 1.

```python
def tim_so_keo_con_lai(n, m, k):
# bat dau viet code cua ban vao duoi day
    return n - m * k
# ket thuc code cua ban

# Luu y: khong thay doi phan code duoi day
n = int(input())
m = int(input())
k = int(input())
print(tim_so_keo_con_lai(n, m, k))
```

# Bài 9

Cho 2 tam giác vuông:

- Tam giác vuông thứ nhất có 2 cạnh góc vuông là a và b là các số nguyên dương

- Tam giác vuông thứ hai có 2 cạnh góc vuông là c và d là các số nguyên dương

Hỏi tổng diện tích của hai tam giác là bao nhiêu? Biết công thức tính diện tích tam giác vuông như sau:

dien_tich_tam_giac_vuong = (a x b) / 2 với a và b là độ dài 2 cạnh góc vuông

[Dữ liệu nhập vào]

-	Hai dòng đầu tiên, mỗi dòng là một số tự nhiên, là hai cạnh góc vuông a và b của tam giác vuông thứ nhất.

-	Hai dòng tiếp theo, mỗi dòng là một số tự nhiên, là hai cạnh góc vuông c và d của tam giác vuông thứ hai.

[Dữ liệu xuất ra]

-	Một số thực, là tổng diện tích của hai tam giác với các chiều dài các cạnh được nhập vào.

Ví dụ (Dấu cách trong dữ liệu nhập xuất tương ứng với dấu xuống dòng):

-	Nhập vào: 3 4 1 6

-	Xuất ra: : 9.0

-	Giải thích: Chương trình được yêu cầu tính tổng diện tích hai tam giác vuông có chiều dài hai cạnh góc vuông là 3, 4 và 1, 6. Tổng diện tích hai tam giác được tính là 3 x 4 / 2 + 1 x 6 / 2 = 9.0. Chương trình xuất ra màn hình giá trị 9.0.

```python
def tinh_tong_dien_tich_tam_giac(a, b, c, d):
# bat dau viet code cua ban vao duoi day
    return a * b / 2 + c * d / 2
# ket thuc code cua ban

# Luu y: khong thay doi phan code duoi day
a = int(input())
b = int(input())
c = int(input())
d = int(input())
tong_dien_tich_tam_giac = tinh_tong_dien_tich_tam_giac(a, b, c, d)
print(tong_dien_tich_tam_giac)
```

# Bài 10

Một công ty sơn tường nhận được yêu cầu sơn một bức tường hình chữ nhật. Viết chương trình giúp công ty tính giá sơn bức tường trên. Biết với mỗi m2 cần sơn, khách hàng cần phải trả 20000 VND. Thông tin cần nhập vào là hai số nguyên bao gồm chiều dài và chiều rộng của bức tường với đơn vị là m.

[Dữ liệu nhập vào]

-	Hai dòng liên tiếp là hai số tự nhiên, lần lượt là chiều dài và chiều rộng của bức tường.

[Dữ liệu xuất ra]

-	Một số tự nhiên là số tiền cần phải trả để sơn bức tường có chiều dài và chiều rộng được nhập vào.

Ví dụ (Dấu cách trong dữ liệu nhập xuất tương ứng với dấu xuống dòng):

-	Nhập vào: 20 10

-	Xuất ra: 4000000

-	Giải thích: Chương trình được yêu cầu tính tiền sơn bức tường có chiều dài và chiều rộng lần lượt là 20m và 10m. Diện tích bức tường là 20 x 10 = 200 m2. Số tiền để sơn xong bức tường là 200 x 20000 = 4000000 VND. Chương trình xuất ra màn hình giá trị 4000000.

```python
def tinh_gia_son_tuong(chieu_dai, chieu_rong):
# bat dau viet code cua ban vao duoi day
    return chieu_dai * chieu_rong * 20000
# ket thuc code cua ban

# Luu y: khong thay doi phan code duoi day
chieu_dai = int(input())
chieu_rong = int(input())
tong_gia_son_tuong = tinh_gia_son_tuong(chieu_dai, chieu_rong)
print(tong_gia_son_tuong)
```
