# Bài 2

Cho 2 số nguyên a và b. (a và b đều khác 0).

Viết chương trình tính bội chung nhỏ nhất của a và b. Xuất kết quả ra màn hình với a và b được nhập vào từ bàn phím.

Nhắc lại: Trong số học, bội số chung nhỏ nhất của hai số nguyên a và b là số nguyên dương nhỏ nhất chia hết cho cả a và b.

Lưu ý: Định dạng kết quả xuất ra phải ở dạng số nguyên.

```python
a = int(input())
b = int(input())
def ucln(a,b):
    if a < b:
        return ucln(b,a)
    while b != 0:
        r = a % b
        a = b
        b = r
    return a

print(int((a*b)/ucln(a,b)))
```

# Bài 3

Viết chương trình xuất ra màn hình tổng của các số lẻ bé hơn n với n > 0 được nhập vào từ bàn phím như code mẫu.

Ví dụ: n = 11. Có 5 số lẻ nhỏ hơn 11 là: 1, 3, 5, 7, 9. Tổng của chúng: 1 + 3 + 5 + 7 + 9. Kết quả cần xuất ra: 25.

```python
n = int(input())
tong = 0
for i in range(n):
    if i % 2 == 1:
        tong += i
print(tong)
```

# Bài 5

Cho s là 1 chuỗi bất kỳ được nhập vào từ bàn phím như code mẫu. Trong chuỗi s, số ký tự là chữ cái xuất hiện nhiều hơn số ký tự là số hay không? Xuất kết quả ra màn hình là: True nếu số lượng chữ cái xuất hiện nhiều hơn. Ngược lại, nếu số chữ cái xuất hiện bé hơn hoặc bằng số lượng ký tự là số thì xuất ra False.

Ví dụ:

- Cho s = "123 BUOC deu Bu@c 1234 !!!".

Chuỗi s gồm có 10 ký tự là chữ cái xuất hiện, lần lượt là: B, U, O, C, d, e, u, B, u, c.

Chuỗi s gồm có 7 ký tự là số xuất hiện, lần lượt là: 1, 2, 3, 1, 2, 3, 4.

Kết quả cần xuất ra màn hình: True
   
- s = "May-2018"

Chuỗi s gồm có 3 ký tự là chữ cái xuất hiện, lần lượt là: M, a, y.

Chuỗi s gồm có 4 ký tự là số xuất hiện, lần lượt là: 2, 0, 1, 8.

Kết quả cần xuất ra màn hình: False

```python
s = input()
chucai = 0
kytu = 0
for i in s:
    if i.isalpha():
        chucai += 1
    if i.isnumeric():
        kytu += 1
print(chucai>kytu)
```

Mỗi người tham gia được phát một cây súng nhựa có thể bắn một lần. Cứ mỗi con robot đổ, nó sẽ làm cho hai con bên cạnh nó đổ theo, 2 con robot bị kéo theo đó lại kéo theo những con kế bên nó (nếu có), quá trình kéo theo cứ thế tiếp tục nếu vẫn còn robot nằm kề robot vừa đổ. Bạn là một trong những người tham gia trò chơi đó. Hãy tính xem nếu bắn trúng, bạn có thể làm đổ tối đa bao nhiêu con Robot.

Tại mỗi vị trí trong hàng, 2 danh hài có thể đặt robot vào hoặc để trống.

Cho chuỗi a biểu diễn trạng thái của hàng đặt robot, xét tại vị trí i trong chuỗi a, nếu a[i] = '1' nghĩa là có đặt robot ở vị trí đó, a[i] = '0' nghĩa là không có robot ở vị trí đó.

Xuất ra màn hình: Số con robot tối đa có thể hạ gục với chuỗi a được nhập vào từ bàn phím như code mẫu.

Ví dụ:



Ở vị dụ trên ta có: a = "10111101"

Tức là tại vị trí 1 và vị trí 6, không có con robot nào.

Tại vị trí 0, 2, 3, 4, 5, 7 có đặt robot.

Nếu ta bắn vào robot thứ 0, thì chỉ đổ 1 robot.

Nhưng nếu giả sử ta bắn vào robot thứ 3, thì nó sẽ kéo theo robot thứ 2 và 4 cùng đổ, robot thứ 4 đổ sẽ kéo theo robot thứ 5 đổ. Vậy tổng cộng có 4 robot đổ.

Kết quả xuất ra màn hình: 4

```python
a = input()
m = []
s = 0
for i in range(len(a)):
    if a[i] == "0":
        m.append(s)
        s = 0
    if a[i] == "1":
        s += 1
    m.append(s)
print(max(m))
```

Batman đang tham gia một nhiệm vụ bí mật. Anh đang chuẩn bị cho kế hoạch đột nhập vào hang ổ tội phạm.
